```python
words = ["WFIDSPQZ", "XHO' 'TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]
import random
word = random.choice(words)
print("Guess a character (uppercase letters only)")
guesses = ''
turns = 10
def decrypt(plaintext):

  ciphertext = ""
  for i in range(len(plaintext)):
    if i % 2 == 0:
      new_even = ord(plaintext[i]) - 1 
      ciphertext += chr(new_even)
    else:
      new_odd = ord(plaintext[i]) + 1
      ciphertext += chr(new_odd)
  return ciphertext

while turns > 0:
  
   failed = 0
   progress = []

   for char in word:
     
       if char in guesses:
           progress.append(char)
           progress.append(" ")
       else:
           progress.append("_ ")
        
           failed += 1
   print(*progress)
   if failed == 0:
       print("You Win")
       print(decrypt("Z@Z"))
      
    
       print("The word is: ", decrypt(word))
       break
  
   
   guess = input("guess a character:")
  

   guesses += guess

   if guess not in word:
      
       turns -= 1
       print("Wrong")
       
      
     
       print("You have ", + turns, ' more guesses')
       if turns == 0:
           print("\nYou Loose")
           print("The word is:", decrypt(word))
           print(decrypt(";'"))
```
